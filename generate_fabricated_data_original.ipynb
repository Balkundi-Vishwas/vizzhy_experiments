```python
df = pd.read_csv(r"c:\Users\Vishwas\Documents\fabricated_real_food.csv")
```

```python
df.columns
```

```python
df['Meal Type'] = df.iloc[:, 0].fillna(method='ffill')  # First column contains meal types
df['Meal Date'] = df.iloc[:, 1].fillna(method='ffill')       # Second column contains dates
df['Meal Time'] = df.iloc[:, 2].fillna(method='ffill')       # Third column contains times

cols = ['Meal Type', 'Meal Date', 'Meal Time'] + [col for col in df.columns if col not in ['Meal Type', 'Meal Date', 'Meal Time']]
df_restructured = df[cols]

```

```python
df_restructured.head()
```

```python
def convert_to_24hr(time_str):
    if pd.isna(time_str):
        return time_str
    try:
        # Parse the time string and convert to 24-hour format
        return datetime.strptime(time_str, '%I:%M %p').strftime('%H:%M')
    except:
        return time_str

df_restructured['time_24hr'] = df_restructured['Meal Time'].apply(convert_to_24hr)
```

```python
df_restructured.head()
```

```python
result = []
for _, row in df_restructured.iterrows():
    # iso_datetime = datetime.strptime(row['DateTime'], "%d-%m-%Y %H:%M").isoformat() + ".000Z"
    
    record = {
        "mealType": row["Meal Type"],
        "mealTime": "01-09-2024" + " "+row["time_24hr"],
        "foodName": row["Name of the Food"],
        "nutrition": {
             "Protein": str(row["Protein (g)"]) +" "+"g",
             "Energy": str(row["Energy (Kcal)"]) +" "+"Kcal",
             "Carbohydrate": str(row["Carbohydrate (g)"]) +" "+"g",
             "Fiber": str(row["Fiber (g)"]) +" "+"g"
             }
    }
    result.append(record)
```

```python
result
```

```python
json_output = json.dumps(result, indent=4)
with open(r"c:\Users\Vishwas\Documents\food_fabricated.json", "w") as outfile:
    outfile.write(json_output)
```
